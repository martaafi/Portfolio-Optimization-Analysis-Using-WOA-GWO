##**C. Stock Sharpe Ratio**

```python
def stock_sharpe_ratio(return_data, rf = 0.06):
  daily_rf = ((1+rf)**(1/236))-1

  mean_return = return_data.mean()
  std_return = return_data.std()
  SR = (mean_return - daily_rf) / std_return

  table_data = {
      'mean_return' : mean_return,
      'standard_deviation' : std_return,
      'sharpe_ratio' : SR
  }

  table = pd.DataFrame(table_data)
  return table
```

```python
tickers = return_df.columns.values.tolist()
sharpe_ratio = stock_sharpe_ratio(return_df[tickers])
sharpe_ratio = sharpe_ratio.reset_index().rename(columns = {'Ticker' : 'ticker'})
sharpe_ratio
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt

# Menentukan jumlah k-cluster optimal dengan silhouette score untuk KMeans
silhouette_avg_kmeans = []

for i in range(2, 10):
    kmeans = KMeans(init='k-means++', n_clusters=i, n_init=50, max_iter=200, random_state=123)
    kmeans.fit(clustering_df)
    silhouette = silhouette_score(clustering_df, kmeans.labels_)
    silhouette_avg_kmeans.append(silhouette)

# Menentukan jumlah k-cluster optimal dengan silhouette score untuk KMedoids
silhouette_avg_kmed = []

for i in range(2, 10):
    kmedoid = KMedoids(n_clusters=i, random_state=26)
    kmedoid.fit(clustering_df)
    silhouette = silhouette_score(clustering_df, kmedoid.labels_)
    silhouette_avg_kmed.append(silhouette)

# Membuat grafik perbandingan silhouette score
plt.figure(figsize=(15, 7))

# Plot untuk KMeans
plt.plot(range(2, 10), silhouette_avg_kmeans, 'bo-', label='KMeans')

# Plot untuk KMedoids
plt.plot(range(2, 10), silhouette_avg_kmed, 'go-', label='KMedoids')

# Menambahkan elemen pada grafik
plt.title('Perbandingan Silhouette Score untuk KMeans dan KMedoids', fontsize=14)
plt.xlabel('Jumlah Cluster (k)', fontsize=12)
plt.ylabel('Silhouette Score', fontsize=12)
plt.xticks(range(2, 10))
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

##**A. Whale Optimization Algorithm**

###**1. High Sharpe Ratio K-Means**

```python
def high_kmeans_optimal_weight(weight):
  def weight_sum(w):
    return 1 - np.sum(w)
  def violate(value):
    return 0 if value < 0 else value
  obj_func =  np.dot(weight.T, np.dot(high_sigma, weight))

  obj_func += violate(weight_sum(weight))
  normalized_weights = weight / np.sum(weight)
  return obj_func
```

```python
from mealpy import FloatVar, WOA

high_kmeans_problem = {
    'obj_func' : high_kmeans_optimal_weight,
    'bounds' : FloatVar(lb = np.zeros(len(high_mean_return)), ub = np.ones(len(high_mean_return))),
    'minmax' : 'min'}
```

```python
optimizer_WOA = WOA.OriginalWOA(epoch = 100, pop_size = 100, pr = 0.05)
optimizer_WOA.solve(high_kmeans_problem, seed = 48)
```

```python
high_woa_kmeans_optimal_weight = []
for i in range(len(high_mean_return)):
  high_woa_kmeans_optimal_weight.append(
      {
          'weight' : optimizer_WOA.g_best.solution[i],
          }
      )
high_woa_kmeans_optimal_weight = pd.DataFrame(high_woa_kmeans_optimal_weight)
high_woa_kmeans_optimal_weight['ticker'] = return_df[highest_kmeans_tickers].columns
high_woa_kmeans_optimal_weight.set_index('ticker', inplace = True)
high_woa_kmeans_optimal_weight
```

##**C. Mean Variance**

```python
def mean_variance_weight(return_df):
  sigma = return_df.cov()
  mean_return = return_df.mean()
  satu = np.full((len(sigma), 1), 1)

  pembilang = np.linalg.inv(sigma)@satu
  penyebut = satu.T@np.linalg.inv(sigma)@satu

  optimal_weight = pd.DataFrame(pembilang/penyebut, columns=['weight'])
  optimal_weight['ticker'] = return_df.columns
  optimal_weight = optimal_weight.set_index('ticker')

  return optimal_weight
```

#**5. Portfolio Performance**

```python
def portfolio_sharpe_ratio(start_date, end_date, weights, rf = 0.06):
  return_df_portfolio = yf.download(weights.index.values.tolist(), start=start_date, end=end_date)['Close'].pct_change().dropna()
  sigma = return_df_portfolio.cov()
  var_rp = np.dot(weights.T, np.dot(sigma, weights))
  e_rp = np.dot(np.transpose(weights), return_df_portfolio.mean())
  E_rp = ((1+e_rp)**30)-1

  std_rp = np.sqrt(var_rp)*np.sqrt(30)
  sharpe_ratio = (E_rp - rf)/std_rp

  result_df = pd.DataFrame({
      'Expected Portfolio Return': [E_rp][0][0],
      'Portfolio Standard Deviation': [std_rp][0][0],
      'Sharpe Ratio': [sharpe_ratio][0][0]})
  return result_df
```

```python
import pandas as pd

date_ranges = [
    ('2024-08-01', '2024-08-31'),
    ('2024-09-01', '2024-09-30'),
    ('2024-10-01', '2024-10-31'),
    ('2024-11-01', '2024-11-30')
]

all_weights = [
    high_kmeans_optimal_weight, high_kmedoid_optimal_weight,
    high_gwo_kmeans_optimal_weight, high_gwo_kmedoid_optimal_weight,
    high_woa_kmeans_optimal_weight, high_woa_kmedoid_optimal_weight
]

all_results = []

method_labels = ['Mean-Var KMeans', 'Mean-Var KMedoid',
                 'GWO KMeans', 'GWO KMedoid',
                 'WOA KMeans', 'WOA KMedoid']

for start_date, end_date in date_ranges:
    results = []
    for weights, method in zip(all_weights, method_labels):
        result_df = portfolio_sharpe_ratio(start_date, end_date, weights)
        result_df['Method'] = method
        results.append(result_df)
    performance_df = pd.concat(results).reset_index(drop=True)
    performance_df['Date Range'] = f'{start_date} - {end_date}'
    all_results.append(performance_df)
```

```python

pd.options.display.float_format = '{:.3f}'.format

final_performance = pd.concat(all_results).reset_index(drop=True)
final_performance = pd.DataFrame(final_performance)
final_performance = final_performance[['Date Range', 'Method', 'Sharpe Ratio']]
final_performance = final_performance.pivot(index = 'Method', columns = 'Date Range', values = 'Sharpe Ratio')
final_performance['Mean'] = final_performance.mean(axis=1)
final_performance['std. Dev'] = final_performance.std(axis=1)
final_performance
```
